## E-Commerce Store EDA

**Scenario**


You are working as a Data Analyst in an e-commerce company. The company has provided you
with customer transaction data. Your task is to analyze the data and generate insights using
NumPy, Matplotlib, and Seaborn. Dataset: ecommerce_customer_spending.csv

**Dataset Columns**

• Customer_ID → Unique ID of customer

• Age → Age of customer

• Gender → Male/Female

• Product_Category → Electronics, Clothing, Grocery, Beauty, Sports

• Purchase_Amount → Amount spent (in PKR)

• Purchase_Date → Date of purchase

• Payment_Mode → Credit Card, Cash, Online Wallet, Bank Transfer


**Part A – NumPy (Statistical Analysis)**

1: Load the dataset using Pandas and convert relevant columns into NumPy arrays.

```python
# Importing the Dependencies required for further parts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the Dataset and observing it
df = pd.read_csv('/content/ecommerce_customer_spending.csv')
df.head()
```

```python
# Converting this dataframe into numpy array
arr = np.asarray(df)
arr
```

2: Find Mean, Median, and Standard Deviation of Purchase_Amount. Maximum and Minimum
purchase amount. Variance in spending behavior.

```python
# Using dataframe to calculate mean
mean = df['Purchase_Amount'].mean()
print("Mean:",mean)

# Using dataframe to calculate median
median = df['Purchase_Amount'].median()
print("Median:",median)

# Using dataframe to calculate standard deviation
standard_deviation = df['Purchase_Amount'].std()
print("Standard Deviation:",standard_deviation)

# Using dataframe to calculate maximum purchase amount
maximum = df['Purchase_Amount'].max()
print("Maximum Purchase Amount:",maximum)

# Using dataframe to calculate minimum purchase amount
minimum = df['Purchase_Amount'].min()
print("Maximum Purchase Amount:",minimum)

# Using dataframe to calculate variance in purchase amount
variance = df['Purchase_Amount'].var()
print("Variance in Purchase Amount:",variance)
```

3: Identify which age group (18–25, 26–35, 36–45, 46–60) spends the most

```python
# Create age groups
bins = [18, 26, 36, 46, 61]
labels = ['18–25', '26–35', '36–45', '46–60']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group by age group and find the group with the maximum total spending
most_spending_age_group = df.groupby('Age_Group')['Purchase_Amount'].sum().idxmax()

print(f"The age group that spends the most is: {most_spending_age_group}")
```

4: Compare average spending of Male vs Female customers.

```python
# Grouping Genders based on avg Purchase Amount
avg_spending_by_gender = df.groupby('Gender')['Purchase_Amount'].mean()
print(avg_spending_by_gender)
```

**Part B – Matplotlib (Visual Insights)**

1: Line Chart → Revenue trend over time.

```python
# Converting the Purchase Date column to date time
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
```

```python
# Plotting the line Plot
plt.figure(figsize=(15, 8))
plt.plot(df['Purchase_Date'], df['Purchase_Amount'], marker='.', markerfacecolor='darkblue', markersize=8)
plt.title("Revenue Trend Over Time")
plt.xlabel("Purchase Date")
plt.ylabel("Total Revenue")
plt.show()
```

2: Bar Chart → Total sales per product category.

```python
# Plotting the Bar Chart
plt.figure(figsize=(10,10))
plt.bar(df['Product_Category'], df['Purchase_Amount'])
plt.title("Total Sales per Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

3: Pie Chart → Percentage contribution of payment modes.

```python
# Plotting the Pie Chart
plt.figure(figsize=(12,8))
plt.pie(df['Payment_Mode'].value_counts(), labels=df['Payment_Mode'].value_counts().index, autopct='%1.1f%%')
plt.legend(title="Payment Mode")
plt.title("Payment Modes by Percentage")
plt.show()
```

4: Scatter Plot → Age vs Purchase Amount.

```python
# Plotting the Scatter Chart
plt.figure(figsize=(12,8))
plt.scatter(x=df['Age'], y=df['Purchase_Amount'])
plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.show()
```

**Part C – Seaborn (Statistical Visualizations)**

1: Histogram → Distribution of purchase amounts.

```python
# Plotting the Histplot
plt.figure(figsize=(12,8))
sns.histplot(df['Purchase_Amount'], kde=True)
plt.title("Distribution of Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()
```

2: Boxplot → Compare spending by gender

```python
# Plotting the Boxplot
plt.figure(figsize=(12,8))
sns.boxplot(x=df['Gender'], y=df['Purchase_Amount'])
plt.title("Comparision of Spending by Gender")
plt.xlabel("Gender")
plt.ylabel("Purchase Amount")
plt.show()
```

3: Heatmap → Correlation between Age, Purchase_Amount, etc

```python
# Plotting the Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

4: Pairplot → Explore relationship between multiple features.

```python
# Plotting the Pair Plot
plt.figure(figsize=(12,8))
sns.pairplot(df)
plt.show()
```

**Part D – Business Insights (Report)**

1: Which product category generates the highest revenue?

```python
highest_reventue_category = df.groupby('Product_Category')['Purchase_Amount'].sum().idxmax()
print(f"The product category that generated the highest revenue is: {highest_reventue_category}")
```

```python
popular_payment_method = df.groupby('Payment_Mode')['Purchase_Amount'].count().idxmax()
print(f"The most popular payment method is: {popular_payment_method}")
```

```python
spending_based_on_age = df.groupby('Age')['Purchase_Amount'].sum().idxmax()
print(f"The most spending age group is: {spending_based_on_age}")
```

```python
spending_based_on_gender = df.groupby('Gender')['Purchase_Amount'].sum().idxmax()
print(f"The most spending gender is of {spending_based_on_gender}")
```
